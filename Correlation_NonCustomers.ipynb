```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\sidharth\Desktop\happycars\NonCustomers.csv')
df.head()
```

```python
df['BirthDate']=pd.to_datetime(df['BirthDate'])
```

```python
df.drop(['NonCustomerID','FirstName','LastName'],'columns', inplace=True)
df.head()
```

```python
df['BirthDate'] = df['BirthDate'].dt.year
df.head()
```

**Label Categorical Data**

```python
replace_map = {'MaritalStatus':{'S':False,'M':True},
'Education':{'Partial High School':0,'High School':1,'Partial College':2,'Bachelors ':3,'Graduate Degree':4},
'Occupation':{'Manual':0,'Clerical':1,'Skilled Manual':2,'Professional':3,'Management':4},
'Gender':{'F':0,'M':1},
'YearlyIncome':{'0-25000':0,'25001-50000':1,'50001-75000':2,'75001-100000':3,'greater than 100000':4},
'CommuteDistance':{'0-1 Miles':0,'1-2 Miles':1,'2-5 Miles':2,'5-10 Miles':3,'10+ Miles':4}}

df.replace(replace_map,inplace=True)
df.head(10)
```

**Correlation Heatmap**

```python
plt.figure(figsize=(16, 9))
heatmap = sns.heatmap(df.corr(), vmin=-1, vmax=1, annot=True)
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':12}, pad=12)
```
